```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
import pandas as pd
df = pd.read_csv("C:\\Users\\USER\\OneDrive\\Desktop\\imdb_top_movies.csv")
df.head()
```

```python
df.shape
df.columns
df.info()
df.describe()
df.isnull().sum()
df = df.dropna()
df["rating"] = df["rating"].astype(float)
df["year"] = df["year"].astype(int)
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["rating"], bins=15, kde=True)
plt.title("IMDb Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,4))
df["year"].value_counts().sort_index().plot()
plt.title("Movies by Release Year")
plt.xlabel("Year")
plt.ylabel("Number of Movies")
plt.show()
plt.figure(figsize=(10,6))
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=df["rank"], y=df["rating"])
plt.title("Rank vs Rating")
plt.xlabel("Rank")
plt.ylabel("Rating")
plt.show()
```
